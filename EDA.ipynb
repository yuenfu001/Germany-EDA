```python
rent_increase = df.groupby("date", as_index=False)[["date","baseRent","totalRent"]].agg("mean").sort_values("date",ascending=False)
rent_increase
```

```python
mapping = {
    "Sep18":0, "May19":1, "Oct19":2, "Feb20":3
}
rent_increase["sorting"] = rent_increase["date"].apply(lambda x: mapping[x])
print(rent_increase)
```

```python
rent_increase_sorted = rent_increase.sort_values("sorting", ascending=True)
rent_increase_sorted
```

```python
xpos = np.arange(len(rent_increase_sorted["date"]))
xpos
```

```python
plt.xticks(xpos, rent_increase_sorted["date"])
plt.ylabel("Amount")
plt.title("Rent Increment with Time")
plt.bar(xpos- 0.2, rent_increase_sorted["baseRent"], label="Base Rent", width=0.4, color="orange")
plt.legend()
plt.bar(xpos+ 0.2,rent_increase_sorted["totalRent"],label="Base Rent",width=0.4,)
plt.legend()
```

```python
plt.xticks(xpos, rent_increase_sorted["date"])
plt.ylabel("Amount")
plt.title("Total Rent Increment with Time")
plt.bar(xpos, rent_increase_sorted["totalRent"])
```

```python
expensive_areas_by_baseRent = round(df[["regio1","baseRent"]].groupby("regio1",as_index=False).mean().sort_values("baseRent",ascending=False))
expensive_areas_by_baseRent
```

```python
ypos = np.arange(len(expensive_areas_by_baseRent["regio1"]))
plt.xticks(ypos, expensive_areas_by_baseRent["regio1"], rotation=90)
plt.ylabel("Amount")
plt.title("Most Expensive Areas")
plt.bar(ypos, expensive_areas_by_baseRent["baseRent"])
```

```python
expensive_areas_by_totalRent = round(df[["regio1","totalRent"]].groupby("regio1",as_index=False).mean().sort_values("totalRent",ascending=False))
expensive_areas_by_totalRent
```

```python
ypos = np.arange(len(expensive_areas_by_totalRent["regio1"]))
plt.xticks(ypos, expensive_areas_by_totalRent["regio1"], rotation=90)
plt.ylabel("Amount")
plt.title("Most Expensive Areas")
plt.bar(ypos, expensive_areas_by_totalRent["totalRent"])
```
